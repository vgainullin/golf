```python
import pandas as pd
import json

with open('q_table.json') as fh:
    qt = json.load(fh)


rows = []
for i, (k, v) in enumerate(qt.items()):
    for action, q_ in v.items():
        row = [k, action, q_]
        rows.append(row)
        
```

```python
df = pd.DataFrame(rows, columns=['state', 'action', 'q'])
```
